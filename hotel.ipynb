```python
import pandas as pd

low_memory=False
```

```python
df = pd.read_csv('zomato.csv', encoding = "ISO-8859-1")

df.head(10)
```

```python
df = df.drop_duplicates().reset_index(drop=True)
df['rest_id'] = df.index

df.head()
```

```python
df2 = pd.read_csv('country.csv')

df2.head(20)
```

```python
inner = pd.merge(df,df2)

inner
```

```python
rest_dim = df[['Restaurant Name','Restaurant ID','Cuisines','Average Cost for two','Price range']].reset_index(drop=True)
rest_dim = rest_dim[['Restaurant ID','Restaurant Name','Cuisines','Average Cost for two','Price range']]

rest_dim.head()
```

```python
country_dim = df2[['Country Code','Country']].reset_index(drop=True)
country_dim = country_dim[['Country Code','Country']]

country_dim.head()
```

```python
location_dim = df[['Restaurant ID','City','Address','Locality','Locality Verbose','Longitude','Latitude']].reset_index(drop=True)
location_dim = location_dim[['Restaurant ID','City','Address','Locality','Locality Verbose','Longitude','Latitude']]

location_dim.head()
```

```python
booking_delivery_dim = df[['Restaurant ID','Has Table booking','Has Online delivery','Is delivering now','Switch to order menu']].reset_index(drop=True)
booking_delivery_dim = booking_delivery_dim[['Restaurant ID','Has Table booking','Has Online delivery','Is delivering now','Switch to order menu']]

booking_delivery_dim.head()
```

```python
rating_dim = df[['Restaurant ID','Aggregate rating','Rating color','Rating text','Votes']].reset_index(drop=True)
rating_dim = rating_dim[['Restaurant ID','Aggregate rating','Rating color','Rating text','Votes']]

rating_dim.head()
```

```python
fact_table = df.merge(rest_dim, on='Restaurant ID') \
             .merge(location_dim, on='Restaurant ID') \
             .merge(booking_delivery_dim, on='Restaurant ID') \
             .merge(rating_dim, on='Restaurant ID') \
             .merge(country_dim, on='Country Code') \
             [['rest_id','Country Code', 'Restaurant ID']]
```
